```python
#Bài thực hành 2,3
import numpy as np
#ma trận phân kỳ
#ma trận A
A = np.array([[0, 1], [1, 0]])
print("Ma trận A:")
print(A)
temp = A.dot(A)
print("\nA^2:")
print(temp)
k = 6
for i in range(k - 1):
    temp = temp.dot(A)
    print(f"\nA^{i+3}:")
    print(temp)
    print()

#ma trận B
B = np.array([[0, -1], [-1, 0]])
print("Ma trận B:")
print(B)
temp = B.dot(B)
print("\nB^2:")
print(temp)
k = 5
for i in range(k - 1):
    temp = temp.dot(B)
    print(f"\nB^{i+3}:")
    print(temp)
    print()

#ma trận hội tụ C
C = np.array([
    [1, 0, 0],
    [0.5, 1, 0],
    [0, 0.5, 1]
])
print("Ma trận C:")
print(C)
temp = C.dot(C)
print("\nC^2:")
print(temp)
k = 1000
for i in range(k - 1):
    temp = temp.dot(C)

print("\nC^1000:")
print(temp)
for i in range(k - 1):
    temp = temp.dot(C)
print("\nC^2000:")
print(temp)


```

```python
#Bài thực hành 3.2
import numpy as np
from scipy import linalg
M = np.array([
    [0.8, 0.3],
    [0.2, 0.7]
])
print("Ma trận M:")
print(M)
print()
P, L, U = linalg.lu(M)

print("ma trận P:")
print(P)
print()

print("ma trận L:")
print(L)
print()

print("ma trận U:")
print(U)
print()

print("L.dot(U):")
print(L.dot(U))
print()
print("nhận xét: kết quả của L.dot(U) (sau hoán vị nếu cần thiết với P) khớp với M")
```

```python
#Bài 5
import numpy as np
A = np.array([
    [0, 1, 0, 1],
    [0, 0, 1, 0],
    [1, 0, 0, 1],
    [1, 1, 0, 0]
])
#tính A^2
A2 = A.dot(A)
print("A^2:")
print(A2)
#tính A^3
A3 = A2.dot(A)
print("\nA^3:")
print(A3)
#tổng A + A^2 + A^3
sumA = A + A2 + A3
print("\nA + A^2 + A^3:")
print(sumA)
k = 3
sumAk = A.copy()
temp = A.copy()

for i in range(1, k):
    temp = temp.dot(A)
    sumAk = sumAk + temp
print(f"\nTổng A + A^2 + ... + A^{k}:")
print(sumAk)
```

```python
#Bài 1
import numpy as np
#a/ tạo ma trận khả nghịch
A = np.array([[2, 1, 1],
              [1, 2, 1],
              [1, 1, 2]])
print("ma trận A:")
print(A)
if np.linalg.det(A) == 0:
    print("ma trận không khả nghịch")
    exit()
else:
    print("ma trận khả nghịch")

#b/ nhập chuỗi cần mã hóa
text = input("nhập tên hoặc mssv: ")

#c/ mã hóa chuỗi
data = [ord(c) for c in text]
while len(data) % 3 != 0:
    data.append(0)

blocks = np.array(data).reshape(-1, 3).T
encoded = A @ blocks
print("mã hóa:")
print(encoded)

#d/ giải mã
A_inv = np.linalg.inv(A)
decoded = np.round(A_inv @ encoded).astype(int)
flat = decoded.T.flatten()
result = ''.join(chr(x) for x in flat if x != 0)
print("giải mã:")
print(result)

```

```python
#Bài 2
#a
def compute_pq(c):
    M = [[1, 0], [0, 1]]

    for ci in c:
        Mi = [[ci, 1], [1, 0]]
        M = [
            [M[0][0]*Mi[0][0] + M[0][1]*Mi[1][0], M[0][0]*Mi[0][1] + M[0][1]*Mi[1][1]],
            [M[1][0]*Mi[0][0] + M[1][1]*Mi[1][0], M[1][0]*Mi[0][1] + M[1][1]*Mi[1][1]]
        ]
    p_n = M[0][0]
    q_n = M[1][0]
    return p_n, q_n

#b
def matrix_mult(A, B):
    return [
        [A[0][0]*B[0][0] + A[0][1]*B[1][0], A[0][0]*B[0][1] + A[0][1]*B[1][1]],
        [A[1][0]*B[0][0] + A[1][1]*B[1][0], A[1][0]*B[0][1] + A[1][1]*B[1][1]]
    ]
def compute_matrix_product(c):
    M = [[1, 0], [0, 1]]
    for ci in c:
        Mi = [[ci, 1], [1, 0]]
        M = matrix_mult(M, Mi)
    return M
def compute_vector_result(M):
    p_n = M[0][0]*1 + M[0][1]*0
    q_n = M[1][0]*1 + M[1][1]*0
    return p_n, q_n
```
